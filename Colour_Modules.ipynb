```python
import cv2

import matplotlib.pyplot as plt

import numpy as np


img=cv2.imread("ash.png")

bgri= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

cmyi = 255- bgri # convert bgr image into cmy image


plt.figure(figsize=(10,10))

plt.subplot(1,2,1)

plt.title('RGB')

plt.imshow(bgri)

plt.axis('off')

plt.subplot(1,2,2)

plt.title('CMYK')

plt.imshow(cmyi)

plt.axis('off')

plt.show()
```

```python

#RGB TO HSV
hsvi=cv2.cvtColor(bgri,cv2.COLOR_RGB2HSV)
h,s,v = cv2.split(hsvi)
hsi=cv2.merge((h,s,v))
hsrg=cv2.cvtColor(hsi,cv2.COLOR_HSV2RGB)

def display_images(h, s, v, bgri, hsvi, hsrg):
    plt.figure(figsize=(13,10))
    for i, (image, title, cmap) in enumerate(zip(images, titles, cmaps), 1):
        plt.subplot(3, 3, i)
        plt.imshow(image, cmap=cmap)
        plt.title(title)
        plt.axis('off')
    plt.show()

images = [h, s, v, h, s, v, bgri, hsvi, hsrg]
titles = ["Hue Channel", "Saturation Channel", "Value Channel","Hue Channel in gray", "Saturation Channel in gray", "Value Channel in gray","RGB image", "HSV image", "Final image"]
cmaps = [None, None, None, 'gray', 'gray', 'gray', None, None, None]
display_images(h, s, v, bgri, hsvi, hsrg)
```

```python
image = bgri
red_image = image.copy()
red_image[:, :, 1] = 0  # Set green channel to 0
red_image[:, :, 2] = 0  # Set blue channel to 0

green_image = image.copy()
green_image[:, :, 0] = 0
green_image[:, :, 2] = 0

blue_image = image.copy()
blue_image[:, :, 0] = 0
blue_image[:, :, 1] = 0

def display_images(images, titles, figsize=(10, 10)):
    plt.figure(figsize=figsize)
    for i, (img, title) in enumerate(zip(images, titles), 1):
        plt.subplot(1, 4, i)
        plt.imshow(img)
        plt.title(title)
        plt.axis('off')
    plt.show()
images = [red_image, green_image, blue_image, bgri]
titles = ["Image with Red Color", "Image with Green Color", "Image with Blue Color", "The RGB image"]
display_images(images, titles)
```
